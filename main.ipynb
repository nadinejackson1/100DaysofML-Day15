```python
# import required libraries
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# download nltk data
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# define sample text for preprocessing
sample_text = """Natural Language Processing is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human language, in particular how to program computers to process and analyze large amounts of natural language data."""
```

```python
# Word tokenization: split text into words
words = word_tokenize(sample_text)
print("Words:", words)

# sentence tokenization: split text into sentences
sentences = sent_tokenize(sample_text)
print("Sentences:", sentences)
```

```python
# remove stopwords
stop_words = set(stopwords.words("english"))

filtered_words = [word for word in words if word.lower() not in stop_words]
print("Filtered words:", filtered_words)
```

```python
# reduce words to their root form
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("Stemmed words:", stemmed_words)
```

```python
# visualize results with bar chart of word frequencies
word_frequencies = nltk.FreqDist(stemmed_words)

plt.figure(figsize=(12, 6))
word_frequencies.plot(30, cumulative=False)
plt.show()
```

```python
# unit test to ensure code words correctly
def test_preprocessing():
    assert len(words) > 0, "Word tokenization failed"
    assert len(sentences) > 0, "Sentence tokenization failed"
    assert len(filtered_words) < len(words), "Stopword removal failed"
    assert len(stemmed_words) == len(filtered_words), "Stemming failed"
print("Test Completed, code is working correctly")
    
test_preprocessing()
```
